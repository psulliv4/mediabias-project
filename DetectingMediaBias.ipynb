# Detecting Media Bias
#### Background and ethical considerations
A growing concern in our society is the prevalence of news media bias. You often hear the allegation that an individual’s viewpoint and political leaning can be traced to the news source that they rely on the most. This can be  troublesome because of those citizens who are susceptible to believing in ‘fake’ or otherwise dishonest news sources. It’s perhaps not possible for a truly unbiased news source to exist, but helping audiences understand that certain news sources may color information in a certain way can help inform viewers of their need to think critically about the decision. We aimed create a classification model can identify the political leanings of a news source. We identified the political leaning of news sources using [The Media Bias Chart](https://www.adfontesmedia.com/?v=402f03a963ba)

The ethical considerations of this project primarily revolve around the ethics of scraping data using the News API, and the effect out research may have on those using the various news sources. Some news agencies do not want people to scrape their data. However, the News API works well on many sites that disallow web crawlers. Therefore, either the sites allow this API or the API fools the sites. No matter the answer to this question, we don't think there is anything unethical about gathering public news from the sources without paywalls. When it comes to those that could be affected, we view this process as educational even if there may be strong feelings involved. Therefore, we view this interaction as ethical as well.

## Data Collection

The first step in our process was to scrape the news articles.  We decided to use News API, because it covered a large range of sources and was free to use for our purposes.  One drawback was that it only allowed us to scrape articles in the past 30 days.  
We used the Media Bias chart created by Ad Fontes Media to determine which sources to pull from.  We gathered data from four left-leaning news sites and four right-leaning news sites. In total, we were able to obtain 3,494 articles from the eight sites.

# Cleaning the Data

```python
import pandas as pd
import numpy as np
import requests
import re
import nltk
import matplotlib.pyplot as plt 
from wordcloud import WordCloud

# Read in data
MarchNews = pd.read_csv('newsapi_scrape_final.csv')

# Replace null values
Nan = MarchNews['content'].isnull().sum()
print(f'There are {Nan} news articles that do not have content')


# Replace Null values with 'None'
MarchNews['content'].fillna("None", inplace = True)
MarchNews['title'].fillna('None', inplace = True)
MarchNews['description'].fillna('None', inplace = True)
```

Next, we found all of the distinct words in the content, description, and title columns. Using the re.findall function we were also able to remove punctuation. After combing through the data we realized that the last three words were for the number of characters after whatever we had and a partial word, so we scrapped the last three entries in the list.

```python
from nltk.corpus import stopwords
stopwords = nltk.corpus.stopwords.words('english')

# Initialize empty lists
Content = []
Description = []
Title = []

# Collect words in each column
for i in range(len(MarchNews['content'])):
    Content.append(re.findall(r'\w+', MarchNews['content'][i])[:-3])
    Description.append(re.findall(r'\w+', MarchNews['description'][i]))
    Title.append(re.findall(r'\w+', MarchNews['title'][i]))
```

Next, we removed stopwords, numbers, and unknown symbols.

```python
# Iterate through df
for i in range(len(Content)):
    
    # Remove stopwords
    Content[i] = [words.lower() for words in Content[i] if words.lower() not in stopwords]
    Title[i] = [words.lower() for words in Title[i] if words.lower() not in stopwords]
    Description[i] = [words.lower() for words in Description[i] if words.lower() not in stopwords]
    
    # Remove non-string values
    Content[i] = [words for words in Content[i] if words.isalpha() == True]
    Title[i] = [words for words in Title[i] if words.isalpha() == True]
    Description[i] = [words for words in Description[i] if words.isalpha() == True]
    
#print(Title)
```

```python
# Assign cleaned lists to columns in df
MarchNews.insert(8,'TitleCleaned', Title)
MarchNews.insert(9,'DescriptionCleaned', Description)
MarchNews.insert(10,'ContentCleaned', Content)
#print(MarchNews[0:2])
```

We then created a column of unique words in each article and a column of all words in each article.

```python
# Get unique words for each column
uniqueWords = []
allWords = []

for i in range(len(MarchNews['ContentCleaned'])): 
    
    # Create empty list each iteration
    lst = []  
    
    # Get all words in title, content, and description
    [lst.append(words) for words in MarchNews['TitleCleaned'][i]]
    [lst.append(words) for words in MarchNews['ContentCleaned'][i]]
    [lst.append(words) for words in MarchNews['DescriptionCleaned'][i]] 
    
    # Assign all words list to each row of whole df
    allWords.append(lst)
    
    # Transform list into set of unique words
    unique = set(lst)
    
    # Assign unique word set to each row of whole df
    uniqueWords.append(unique)
    
# Create new columns using uniqueWords and allWords lists
MarchNews.insert(11,'uniqueWords', uniqueWords)
MarchNews.insert(12, 'allWords', allWords)
MarchNews[1:2]
```

Next, we created column of all words in an article saved as a single string. This will be fed into the ML algorithms.

```python
# Initialize empty list
allWords_string = []

for i in range(len(MarchNews['ContentCleaned'])):
    
    # Set allWords as a single string
    word_list = MarchNews['allWords'][i]
    word_string =  (' '.join(word for word in word_list))
    allWords_string.append(word_string)

# Create new column from data above
MarchNews.insert(13, 'allWords_sentenceString', allWords_string)
MarchNews[1:3]
```

## Classifiers
We decided to test four classifiers on our data. We used logistic regression, multi-level percenptron, naive bayes, and SVM. Before we could run the classifiers, we vectorized the data and split it into training and test groups.

```python
from sklearn.model_selection import train_test_split, cross_val_predict, cross_val_score, KFold
from sklearn import svm, metrics
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
plt.style.use('ggplot')

# Get data and targets
X = MarchNews['allWords_sentenceString'].tolist() # All words/data
y = np.asarray(MarchNews['bias'])  # Target

# Vectorizers 
vectorizer_TF = TfidfVectorizer(min_df = 2, max_df = .6)
TFVectorizer_X = vectorizer_TF.fit_transform(X)

vectorizer_CV = CountVectorizer(min_df = 2, max_df = .6)
CVectorizer_X = vectorizer_CV.fit_transform(X)

# Split into train and test data (TfidfVectorizer, CountVectorizer, and homemade)
Xtrain_TF, Xtest_TF, ytrain_TF, ytest_TF = train_test_split(TFVectorizer_X, y, random_state=1, test_size=0.8)
Xtrain_CV, Xtest_CV, ytrain_CV, ytest_CV = train_test_split(CVectorizer_X, y, random_state=1, test_size=0.8)
Xtrain_HomeGrown, Xtest_HomeGrown, ytrain_HG, ytest_HG = train_test_split(wordCountMatrix, y, random_state=1, test_size=0.8)
```

```python
# Get bigram vectors

# TfidfVectorizer
vectorizer_TF_bi = TfidfVectorizer(min_df = 2, max_df = .6, ngram_range=(2,2))
TFVectorizer_X_bi = vectorizer_TF_bi.fit_transform(X)

# CountVectorizer
vectorizer_CV_bi = CountVectorizer(min_df = 2, max_df = .6, ngram_range=(2,2))
CVectorizer_X_bi = vectorizer_CV_bi.fit_transform(X)

# Split into train and test data
Xtrain_TF_bi, Xtest_TF_bi, ytrain_TF_bi, ytest_TF_bi = train_test_split(TFVectorizer_X_bi, y, random_state=1, test_size=0.8)
Xtrain_CV_bi, Xtest_CV_bi, ytrain_CV_bi, ytest_CV_bi = train_test_split(CVectorizer_X_bi, y, random_state=1, test_size=0.8)
```

```python
# Get trigtam vectors

# TfidfVectorizer
vectorizer_TF_tri = TfidfVectorizer(min_df = 2, max_df = .6, ngram_range=(3,3))
TFVectorizer_X_tri = vectorizer_TF_tri.fit_transform(X)

# CountVectorizer
vectorizer_CV_tri = CountVectorizer(min_df = 2, max_df = .6, ngram_range=(3,3))
CVectorizer_X_tri = vectorizer_CV_tri.fit_transform(X)

# Split into train and test data
Xtrain_TF_tri, Xtest_TF_tri, ytrain_TF_tri, ytest_TF_tri = train_test_split(TFVectorizer_X_tri, y, random_state=1, test_size=0.8)
Xtrain_CV_tri, Xtest_CV_tri, ytrain_CV_tri, ytest_CV_tri = train_test_split(CVectorizer_X_tri, y, random_state=1, test_size=0.8)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
print('Logistic Regression')

# Unigrams
# Using TfidfVectorizer
LR_TF = LogisticRegression(random_state=0).fit(Xtrain_TF, ytrain_TF)
y_TF = LR_TF.predict(Xtest_TF)
print('\nAccuracy using TfidfVectorizer =', metrics.accuracy_score(y_true = ytest_TF, y_pred = y_TF))

# Using CountVectorizer
LR_CV = LogisticRegression(random_state=0).fit(Xtrain_CV, ytrain_CV)
y_CV = LR_TF.predict(Xtest_CV)
print('Accuracy using CountVectorizer =', metrics.accuracy_score(y_true = ytest_CV, y_pred = y_CV))

# Using HomeGrown Matrix
LR_HG = LogisticRegression(random_state=0).fit(Xtrain_HomeGrown, ytrain_HG)
y_HG = LR_HG.predict(Xtest_HomeGrown)
print('Accuracy using a homemade vectorizer =', metrics.accuracy_score(y_true = ytest_HG, y_pred = y_CV))


# Bigrams
# Using TfidfVectorizer
LR_TF = LogisticRegression(random_state=0).fit(Xtrain_TF_bi, ytrain_TF_bi)
y_TF = LR_TF.predict(Xtest_TF_bi)
print('\nAccuracy using TfidfVectorizer bigrams =', metrics.accuracy_score(y_true = ytest_TF_bi, y_pred = y_TF))

# Using CountVectorizer
LR_CV = LogisticRegression(random_state=0).fit(Xtrain_CV_bi, ytrain_CV_bi)
y_CV = LR_TF.predict(Xtest_CV_bi)
print('Accuracy using CountVectorizer bigrams =', metrics.accuracy_score(y_true = ytest_CV_bi, y_pred = y_CV))


# Trigrams
# Using TfidfVectorizer
LR_TF = LogisticRegression(random_state=0).fit(Xtrain_TF_tri, ytrain_TF_tri)
y_TF = LR_TF.predict(Xtest_TF_tri)
print('\nAccuracy using TfidfVectorizer Trigrams =', metrics.accuracy_score(y_true = ytest_TF_tri, y_pred = y_TF))

# Using CountVectorizer
LR_CV = LogisticRegression(random_state=0).fit(Xtrain_CV_tri, ytrain_CV_tri)
y_CV = LR_TF.predict(Xtest_CV_tri)
print('Accuracy using CountVectorizer Trigrams =', metrics.accuracy_score(y_true = ytest_CV_tri, y_pred = y_CV))
```

```python
# Multi-layer Perceptron
from sklearn.neural_network import MLPClassifier
print('Cross Validated Multi-layer Perceptron')

# Get itorater list for cross-validation
rng = np.linspace(0.0001,0.01,110)

# Create lists with 0s the same length of the itorater
accuracies_mlp_TF = np.zeros(rng.shape[0])
accuracies_mlp_TF_bi = np.zeros(rng.shape[0])
accuracies_mlp_TF_tri = np.zeros(rng.shape[0])

accuracies_mlp_CV = np.zeros(rng.shape[0])
accuracies_mlp_CV_bi = np.zeros(rng.shape[0])
accuracies_mlp_CV_tri = np.zeros(rng.shape[0])

t1_loop = process_time() # Start time

# Find accuracies at each itorator
for i,a in enumerate(rng): 
    
    # Train models
    MLP_TF = MLPClassifier(solver='lbfgs', alpha=a, hidden_layer_sizes=(100, 100), random_state=1, max_iter = 1000).fit(Xtrain_TF, ytrain_TF)
    MLP_TF_bi = MLPClassifier(solver='lbfgs', alpha=a, hidden_layer_sizes=(100, 100), random_state=1, max_iter = 1000).fit(Xtrain_TF_bi, ytrain_TF_bi)
    MLP_TF_tri = MLPClassifier(solver='lbfgs', alpha=a, hidden_layer_sizes=(100, 100), random_state=1, max_iter = 1000).fit(Xtrain_TF_tri, ytrain_TF_tri)
    
    MLP_CV = MLPClassifier(solver='lbfgs', alpha=a, hidden_layer_sizes=(100, 100), random_state=1, max_iter = 1000).fit(Xtrain_CV, ytrain_CV)
    MLP_CV_bi = MLPClassifier(solver='lbfgs', alpha=a, hidden_layer_sizes=(100, 100), random_state=1, max_iter = 1000).fit(Xtrain_CV_bi, ytrain_CV_bi)
    MLP_CV_tri = MLPClassifier(solver='lbfgs', alpha=a, hidden_layer_sizes=(100, 100), random_state=1, max_iter = 1000).fit(Xtrain_CV_tri, ytrain_CV_tri)
    
    # Validate models 
    y_TF = MLP_TF.predict(Xtest_TF)
    y_TF_bi = MLP_TF_bi.predict(Xtest_TF_bi)
    y_TF_tri = MLP_TF_tri.predict(Xtest_TF_tri)
    
    y_CV = MLP_CV.predict(Xtest_CV)
    y_CV_bi = MLP_CV_bi.predict(Xtest_CV_bi)
    y_CV_tri = MLP_CV_tri.predict(Xtest_CV_tri)
    
    # Store in list
    accuracies_mlp_TF[i]  = metrics.accuracy_score(y_true = ytest_TF, y_pred = y_TF)
    accuracies_mlp_TF_bi[i] = metrics.accuracy_score(y_true = ytest_TF_bi, y_pred = y_TF_bi)
    accuracies_mlp_TF_tri[i] = metrics.accuracy_score(y_true = ytest_TF_tri, y_pred = y_TF_tri)

    accuracies_mlp_CV[i]  = metrics.accuracy_score(y_true = ytest_CV, y_pred = y_CV)
    accuracies_mlp_CV_bi[i] = metrics.accuracy_score(y_true = ytest_CV_bi, y_pred = y_CV_bi)
    accuracies_mlp_CV_tri[i] = metrics.accuracy_score(y_true = ytest_CV_tri, y_pred = y_CV_tri)
    
# Plot model accuracy as alpha changes 
plt.plot(rng,accuracies_mlp_TF)
plt.title("Multi-layer Perceptron Accuracy by alpha (TfidfVectorizer Unigrams)")
plt.show()  
plt.plot(rng,accuracies_mlp_TF_bi)
plt.title("Multi-layer Perceptron Accuracy by alpha (TfidfVectorizer Bigrams)")
plt.show() 
plt.plot(rng,accuracies_mlp_TF_tri)
plt.title("Multi-layer Perceptron Accuracy by alpha (TfidfVectorizer Trigrams)")
plt.show() 

plt.plot(rng,accuracies_mlp_CV)
plt.title("Multi-layer Perceptron Accuracy by alpha (CountVectorizer Unigrams)")
plt.show()  
plt.plot(rng,accuracies_mlp_CV_bi)
plt.title("Multi-layer Perceptron Accuracy by alpha (CountVectorizer Bigrams)")
plt.show() 
plt.plot(rng,accuracies_mlp_CV_tri)
plt.title("Multi-layer Perceptron Accuracy by alpha (CountVectorizer Trigrams)")
plt.show()

# Print results
print ("The optimal alpha using TfidfVectorizer and unigrams exists at " +
       str(round(rng[np.where(accuracies_mlp_TF == (max(accuracies_mlp_TF)))[0][0]], 6)) +", with an accuracy of " +
       str(round(max(accuracies_mlp_TF), 5)))
print ("The optimal alpha using TfidfVectorizer and bigrams exists at " +
       str(round(rng[np.where(accuracies_mlp_TF_bi == (max(accuracies_mlp_TF_bi)))[0][0]], 6)) +", with an accuracy of " +
       str(round(max(accuracies_mlp_TF_bi), 5)))
print ("The optimal alpha using TfidfVectorizer and trigrams exists at " +
       str(round(rng[np.where(accuracies_mlp_TF_tri == (max(accuracies_mlp_TF_tri)))[0][0]], 6)) +", with an accuracy of " +
       str(round(max(accuracies_mlp_TF_tri), 5)))

print ("\nThe optimal alpha using CountVectorizer and unigrams exists at " +
       str(round(rng[np.where(accuracies_mlp_CV == (max(accuracies_mlp_CV)))[0][0]], 6)) +", with an accuracy of " +
       str(round(max(accuracies_mlp_CV), 5)))
print ("The optimal alpha using CountVectorizer and bigrams exists at " +
       str(round(rng[np.where(accuracies_mlp_CV_bi == (max(accuracies_mlp_CV_bi)))[0][0]], 6)) +", with an accuracy of " +
       str(round(max(accuracies_mlp_CV_bi), 5)))
print ("The optimal alpha using CountVectorizer and trigrams exists at " +
       str(round(rng[np.where(accuracies_mlp_CV_tri == (max(accuracies_mlp_CV_tri)))[0][0]], 6)) +", with an accuracy of " +
       str(round(max(accuracies_mlp_CV_tri), 5)))

t2_loop = process_time() # End time
t_tot = t2_loop - t1_loop # Time elapsed
print ('Time elapsed: ' + str(t_tot) + ' seconds')
```
